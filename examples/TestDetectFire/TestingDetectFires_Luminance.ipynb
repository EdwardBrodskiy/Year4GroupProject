# Testing Dectect Fires - Luminance


```python
import cv2
import numpy as np
import glob
```

```python
#Identifies fires in forest fire
def detectFires(image, imName):
    #Convert BGR to YCbCr
    im = cv2.cvtColor(image, cv2.COLOR_BGR2YCR_CB)
    #Extract Y channel (luminance) as own array
    imY = im[:,:,0] 
    #Loop for each pixel (same dimensions for each, doesn't matter which array)
    for i in range(imY.shape[0]):
        for j in range(imY.shape[1]):
            #Check luminance value for pixel (Check in imY as contains Y value)
            if imY[i][j] > 150:
                #Overlay fire pixels with yellow (Changes bgr of original image, dimensions of image and imY correspond to same pixel)
                image[i][j][0] = 0
                image[i][j][1] = 255
                image[i][j][2] = 255
    #Write new image in detectImages
    cv2.imwrite('detectImages/fire{}.jpg'.format(imName), image)
    #Marker for saved image progress
    print('fire',imName)
    

# select the path
path = "images\*.*"
#Run function for each image
for file in glob.glob(path):
    #Pass image and image number 
    detectFires(cv2.imread(file), file[7:-5])
```
